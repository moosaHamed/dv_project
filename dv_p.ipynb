The dataset focuses on two types of maintenance performed on HVAC systems: **routine maintenance** and **reactive maintenance**. Routine maintenance is planned and scheduled on a monthly basis, whereas reactive maintenance is unplanned and conducted in response to faults.
For simplicity, the 'Asset Name' column has been aliased as 'Asset' with a numeric identifier. The 'Location' column has been removed, and 'Asset Model' has also been aliased.

The goal of this data visualization is to identify HVAC brands with the fewest faults, minimal spare parts replacements, and reduced repair frequencies. These insights will help product engineers in selecting reliable HVAC systems for future projects.Following data represent two types of maintenance: routine and reactive for HVAC Systems. Routine maintenance is scheduled and defined before carried in Monthly bases while reactive one carried suddenly when fault occur.

Data are aliased from perspective of Asset Name and Location column removed. Asset Model is aliased as well.
Rhe purpose of following Data Visualization is to study which brands has minimum faults, spare parts replacements, and less repaired times. These visialuzation will help Product Engineers to select HVAC Systems for future projects.

```python
unique_values = np.array(list(set(df_can.iloc[:, 5:].values.flatten())))
print(unique_values)
```

```python
asset_brand_count = df_can.head(12)['Asset Brand'].value_counts()
print(asset_brand_count)
```

```python
unique_brands = np.array(list(set(df_can['Asset Brand'])))
unique_values = np.array(list(set(df_can.iloc[:, 5:].values.flatten())))
print(unique_brands)
print(unique_values)
```

```python
brand_dataframes = {}
for brand in unique_brands:
    brand_df = df_can[df_can['Asset Brand'] == brand].iloc[:, 5:]
    counts = {value: (brand_df == value).sum().sum() for value in unique_values}
    brand_dataframes[brand] = pd.DataFrame([counts], columns=unique_values)

for brand, df in brand_dataframes.items():
    for column in df.columns:
        print(f'{brand}: {column}: {df[column].sum()}')
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
fault_reported_counts = sorted([brand_dataframes[brand]['Fault Reported'].sum() for brand in unique_brands], reverse=True)
repaired_counts = sorted([brand_dataframes[brand]['Repaired'].sum() for brand in unique_brands], reverse=True)
spare_part_counts = sorted([brand_dataframes[brand]['Spare part replaced'].sum() for brand in unique_brands], reverse=True)

width = 0.75
x = np.arange(len(unique_brands))

fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(x, fault_reported_counts, width, label='Fault Reported', color='red')
ax.bar(x, repaired_counts, width, bottom=fault_reported_counts, label='Repaired', color='blue')
ax.bar(x, spare_part_counts, width, bottom=np.array(fault_reported_counts) + np.array(repaired_counts), label='Spare Parts Replaced', color='green')

ax.set_xticks(x)
ax.set_xticklabels(unique_brands, rotation=45, ha='right')
ax.set_ylabel('Count')
ax.set_title('Stacked Bar Chart: Fault Reported, Repaired, and Spare Parts Replaced (Sorted High to Low)')
ax.legend()

# Remove the background grid
ax.grid(False)

plt.tight_layout()
plt.show()
```

Observing that **Samsung, Daikin** and **LG** have less cost in maintenance and proven to be charm choices from other brands.
